```python
#Copy_safty
df_fe = df.copy()
```

```python
#STEP 1: TARGET ENCODING (NO-SHOW)

df_fe['no_show']=df_fe['no_show'].map({'no':0,'yes':1})
df_fe
```

```python
df['no_show'] = df['no_show'].astype(str).str.strip().str.lower()

df['no_show'] = df['no_show'].map({'no': 0, 'yes': 1})
```

```python
df['no_show'].value_counts(dropna=False)
```

```python
#STEP 2: DATE FEATURES (CORE FEATURE)
df_fe['appointment_time'] = pd.to_datetime(df_fe['appointment_time'], errors='coerce')
df_fe['appointment_date_continuous'] = pd.to_datetime(df_fe['appointment_date_continuous'], errors='coerce')

df_fe['hour'] = df_fe['appointment_time'].dt.hour
df_fe['day'] = df_fe['appointment_time'].dt.day
df_fe['month'] = df_fe['appointment_time'].dt.month
df_fe['weekday'] = df_fe['appointment_time'].dt.dayofweek
df_fe['week'] = df_fe['appointment_date_continuous'].dt.isocalendar().week
```

```python
#STEP 3: AGE FEATURES
df_fe['age_group'] = pd.cut(
    df_fe['age'],
    bins=[0, 12, 18, 40, 60, 100],
    labels=['child', 'teen', 'adult', 'middle', 'senior']
)
```

```python
#STEP 4: HEALTH RISK SCORE
df_fe['health_risk'] = (
    df_fe['Hipertension'] +
    df_fe['Diabetes'] +
    df_fe['Alcoholism'] +
    df_fe['Handcap']
)
```

```python
#STEP 5: SUPPORT NEED FEATURE
df_fe['needs_support'] = (
    df_fe['patient_needs_companion'] |
    df_fe['under_12_years_old'] |
    df_fe['over_60_years_old']
)
```

```python
#STEP 6: WEATHER FEATURES
df_fe['high_rain'] = (df_fe['average_rain_day'] > df_fe['average_rain_day'].median()).astype(int)
df_fe['high_temp'] = (df_fe['max_temp_day'] > df_fe['max_temp_day'].median()).astype(int)

df_fe['weather_risk'] = df_fe['high_rain'] + df_fe['high_temp']
```

```python
#STEP 7: INTERACTION FEATURES
df_fe['age_health_interaction'] = df_fe['age'] * df_fe['health_risk']
```

```python
#STEP 8: SPECIALTY FEATURE
high_risk_specialties = [
    'psychotherapy',
    'speech therapy',
    'physiotherapy',
    'occupational therapy'
]

df_fe['specialty_risk'] = df_fe['specialty'].apply(
    lambda x: 1 if x in high_risk_specialties else 0
)
```
