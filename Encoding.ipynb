```python
num_cols = df.select_dtypes(
    include=['int64', 'float64']
).columns

print("Numerical Columns:")
print(num_cols)
```

```python
#Encoding----
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

cols = [
    'specialty',
    'gender',
    'disability',
    'appointment_shift'
]

for col in cols:
    df[col] = le.fit_transform(df[col].astype(str))
```

```python
df['rain_intensity'] = le.fit_transform(df['rain_intensity'].astype(str))
df['heat_intensity'] = le.fit_transform(df['heat_intensity'].astype(str))
```

```python
df.drop([
    'appointment_time',
    'appointment_date_continuous',
   
], axis=1, inplace=True)
```

```python
# Frequency Encoding for 'place'

freq_map = df['place'].value_counts()

df['place_encoded'] = df['place'].map(freq_map)

# Check result
print(df[['place', 'place_encoded']].head())
```

```python
df.drop('place', axis=1, inplace=True)
```

```python
cat_cols = df.select_dtypes(
    include=['object', 'category']
).columns

print("Catogircal Columns:")
print(cat_cols)
```
